### Ist der Wrfel Gezinkt?


```python
import numpy as np
from collections import Counter

# Die Protokollierten Messwerte
Messwerte = [4, 1, 1, 5, 3, 3, 1, 4, 5, 6, 1, 4, 1, 6, 1, 2, 1, 2, 3, 5]
anzahl_messungen = len(Messwerte)
```

```python
# Absolute Hufigkeit ermitteln
absolute_haeufigkeit = Counter(Messwerte)
print("Absolute Hufigkeit")
print(absolute_haeufigkeit)
```

```python
# Relative Hufigkeit ermitteln (relfreq = absolute Hufigkeit / Anzahl Messungen)
from scipy.stats import relfreq

# Die Werte liegen zwischen 1 und 6, also 6 Bins (Klassen)
# numbins gibt die relative hufigkeit pro Zahl aus
ergebnis_relfreq = relfreq(Messwerte, numbins=6)

# die relativen hufigkeiten befinden sich im ersten Element
relative_haeufigkeiten_array =  ergebnis_relfreq[0]

# Erstellen einer Liste der Ergebnisse, um die Ausgabe Klarer zu machen
zahlen = np.arange(1, 7)
print("\nRelative Hufigkeit")
for zahl, rel_haeufigkeit in zip(zahlen, relative_haeufigkeiten_array):
    print(f"{zahl}: {rel_haeufigkeit:.4f} ({rel_haeufigkeit*100:.2f}%)")


```

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# 1. Die Messwerte
messwerte = np.array([4, 1, 1, 5, 3, 3, 1, 4, 5, 6, 1, 4, 1, 6, 1, 2, 1, 2, 3, 5])

# 2. Absolute Hufigkeiten ermitteln (mit Counter)
haeufigkeiten = Counter(messwerte)

# 3. Daten fr den Plot vorbereiten
# Die Wrfelzahlen (x-Achse)
zahlen = sorted(haeufigkeiten.keys())
# Die Anzahl der Male (y-Achse)
anzahlen = [haeufigkeiten[zahl] for zahl in zahlen]

# 4. Plot erstellen
plt.figure(figsize=(8, 5))

# Erzeuge das Balkendiagramm
bars = plt.bar(zahlen, anzahlen, color='skyblue', edgecolor='black')

# Beschriftung und Titel
plt.title('Absolute Hufigkeiten der Wrfelwerte (N=20)')
plt.xlabel('Wrfelzahl')
plt.ylabel('Absolute Hufigkeit (Anzahl der Wrfe)')

# Achsen-Ticks setzen, um nur ganze Zahlen (1 bis 6) anzuzeigen
plt.xticks(zahlen)

# Optional: Die absoluten Werte direkt auf die Balken schreiben
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, int(yval), ha='center', va='bottom')

# Eine erwartete Linie hinzufgen (nur zur visuellen Hilfe)
# Erwartung: 20 Messungen / 6 Zahlen  3.33
erwartete_anzahl = len(messwerte) / 6
plt.axhline(erwartete_anzahl, color='red', linestyle='--', linewidth=1, label=f'Erwarteter Mittelwert ({erwartete_anzahl:.2f})')

plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

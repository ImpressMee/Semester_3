```python
# numpy referenz https://numpy.org/doc/stable/reference/generated/numpy.median.html
import numpy as np
import statistics as stat

urliste = [2,4,3,1,2,4,2,2,2,3]
sortierteliste = sorted(urliste)

print("Sortierte Liste:", sortierteliste)
```

```python
# a) Das Arithmetische Mittel
arithmetisches_mittel = np.mean(sortierteliste)
print("Arithmetisches Mittel:", arithmetisches_mittel)
```

```python
# b) der median
median = np.median(sortierteliste)
print("Der Median ist: ", median)
```

```python
# Um den Modalwert zu berechnen, schaut man sich in der liste an, was am häugisten vor kommt.
# In dem Fall kommt die 2, 5x vor
# c) der modalwert

for wert in sorted(set(sortierteliste)):
    count = sortierteliste.count(wert)
    print(f"Die {wert} kam {count} mal vor")

modalwert = stat.mode(sortierteliste)
print("Der Modalwert ist: ", modalwert)
```

```python
# d) Das 10% Quantil
# Hier wird, per referenz, ein Linear-algorithmus benutzt, weswegen 1.9 raus kommt.
# Man kann auch andere Algorithmen benutzen mit verschiedenen ergebnissen
# In meiner bestimmung kam einfach nur 1 raus. Position = 10 * 0,1 = 1. Die erste Position ist eine 1
q1 = np.percentile(sortierteliste, 10, method="linear") #default, method ist optional
q2 = np.percentile(sortierteliste, 10, method="inverted_cdf")
q3 = np.percentile(sortierteliste, 10, method="averaged_inverted_cdf")
q4 = np.percentile(sortierteliste, 10, method="closest_observation")
q5 = np.percentile(sortierteliste, 10, method="interpolated_inverted_cdf")
q6 = np.percentile(sortierteliste, 10, method="hazen")
q7 = np.percentile(sortierteliste, 10, method="weibull")
q8 = np.percentile(sortierteliste, 10, method="median_unbiased")
q9 = np.percentile(sortierteliste, 10, method="normal_unbiased")

print("Linear Algorithmus: ", q1)
print("Inverted_cdf: ", q2)
print("averaged inverted cdf: ", q3)
print("closest obervation: ", q4)
print("interpolated inverted cdf: ", q5)
print("Hazen: ", q6)
print("weibull: ", q7)
print("median_unbiased: ", q8)
print("normal unbiased: ", q9)

```

d) Die Empirische Standartabweichung
Man berechnet das indem man die Wurzel der empirischen Varianz zieht.

Die Empirische Varianz berechnet man so:


mit





```python
import math
# Aufgabe 1.5
# Aufgabe 1.5.1
urliste = [25, 10, 30, 25, 35, 25]
sortierteliste = sorted(urliste)
arithmetisches_mittel = np.mean(sortierteliste)
der_median = np.median(sortierteliste)
q1 = np.percentile(sortierteliste, 75)

print(f"a) arithmetisches mittel: {arithmetisches_mittel}")
print(f"b) der median ist: {der_median}")
print(f"c) Das 75% Quantil ist: {q1}")

# d) Empirische Standartabweichung
n = len(urliste)

x = 0
print("d) ----------------------")

for werte in range(n):
    # Wenn es nicht das letzte Element ist → Plus drucken
    if werte < n - 1:
        print(f"({sortierteliste[werte]} - {arithmetisches_mittel})² +", end=" ")
    else:
        print(f"({sortierteliste[werte]} - {arithmetisches_mittel})²", end="")

    x += (sortierteliste[werte] - arithmetisches_mittel) ** 2

print(f"\nDas Ergebnis der Summe ist {x}")
s2 = x/n-1 # empirische Varianz
s = math.sqrt(s2) # empirische Standartabweichung

print("-------------------------")
print(f"Empirische Varianz: s2 = ({x} / {n}-1) = {s2}")
print(f"Empirische Standartabweichung:s = sqrt({s2}) = {s}")


```

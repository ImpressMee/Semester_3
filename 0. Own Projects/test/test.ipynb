```python
import matplotlib.pyplot as plt

# Beispiel-Daten
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]

# Plot
plt.plot(x, y, marker="o", linestyle="-", color="b", label="Datenpunkte")

# Achsenbeschriftungen und Titel
plt.xlabel("x-Werte")
plt.ylabel("y-Werte")
plt.title("Einfacher Plot in Python")
plt.legend()

# Anzeigen
plt.show()
```
